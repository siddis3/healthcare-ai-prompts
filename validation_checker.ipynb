```python
# A reusable validation checker
def check_discrepancy(epic, tableau, threshold=10):
    diff = epic - tableau
    if diff > threshold:
        return f"⚠️ FLAG: Difference of {diff} exceeds threshold"
    else:
        return f"✅ PASS: Difference of {diff} is within range"

print(check_discrepancy(156, 142))
print(check_discrepancy(100, 98))
```
